```python
with open('artist_id.txt', 'w') as f:
    for item in ids:
        f.write("%s\n" % item)
```

```python
for i,j in art_genres.items():
    if re.match(".*pop|neo mellow|boy band",j):
        art_genres[i] = re.sub(".*pop|neo mellow|boy band","pop",j)
    elif re.match(".*rap",j):
        art_genres[i] = re.sub(".*rap","rap",j)
    elif re.match(".*rock",j):
        art_genres[i] = re.sub(".*rock","rock",j)
    elif re.match(".*hip hop",j):
        art_genres[i] = re.sub(".*hip hop","hip hop",j)
    elif re.match(".*country",j):
        art_genres[i] = re.sub(".*country","country",j)
    elif re.match(".*r&b|.*soul|afrofuturism",j):
        art_genres[i] = re.sub(".*r&b|.*soul|afrofuturism","r&b",j)
    elif re.match(".*latin",j):
        art_genres[i] = re.sub(".*latin","latin",j)
    elif re.match("brostep|complextro",j):
        art_genres[i] = re.sub("brostep|complextro","electro house",j)
    
    
    
    
```

```python
with open('artist_genres.txt', 'w') as convert_file:
     convert_file.write(json.dumps(art_genres))
```
